```python
import plotly.express as px
from merge2 import main_merge

df = main_merge()

# Map full state names to abbreviations
state_abbrev = {
    "alaska": "AK", "arizona": "AZ", "california": "CA", "colorado": "CO",
    "connecticut": "CT", "florida": "FL", "hawaii": "HI", "idaho": "ID",
    "illinois": "IL", "indiana": "IN", "kansas": "KS", "kentucky": "KY",
    "louisiana": "LA", "maine": "ME", "maryland": "MD", "massachusetts": "MA",
    "michigan": "MI", "minnesota": "MN", "montana": "MT", "nebraska": "NE",
    "nevada": "NV", "new jersey": "NJ", "new mexico": "NM", "new york": "NY",
    "north carolina": "NC", "ohio": "OH", "oklahoma": "OK", "oregon": "OR",
    "pennsylvania": "PA", "rhode island": "RI", "texas": "TX", "virginia": "VA",
    "washington": "WA", "west virginia": "WV", "wisconsin": "WI"
}# had GEM create and utilized lab 09 from sd211 to make this dictionary faste

# Add abbreviation column
df["state_code"] = df["State"].str.lower().map(state_abbrev)#Gem advised to make this new column and helped produce the code for utilizing the dictionary above

fig = px.choropleth(df,locations="state_code",locationmode="USA-states",color="federal_action_obligation",scope="usa",animation_frame="Year", color_continuous_scale="cividis",hover_name="State",hover_data={"Political Majority": True,"Lean": True, "state_code": False})
#used lab09 from sd211 for choropleth idea, GEM helped make interactive with the animation frame to change years, color_continous_scale was both utilized from lab09 and advised by gem
fig.update_layout(title="Federal Action Obligation by State (2020 vs 2024)")

fig.show()
```
